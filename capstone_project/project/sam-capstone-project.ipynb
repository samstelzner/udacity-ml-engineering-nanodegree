```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
bucket = 'sam-capstone-project'

preprocessed_train_path = 'preprocessed-data/train.csv'
preprocessed_test_path = 'preprocessed-data/test.csv'
```

```python
import pandas as pd
```

```python
s3uri_preprocessed_train = 's3://{}/{}'.format(bucket, preprocessed_train_path)
df_train = pd.read_csv(s3uri_preprocessed_train)
```

```python
df_train.head()
```

```python
predictor = TabularPredictor(label="Survived").fit(
    train_data=df_train, time_limit=120, presets="best_quality"
)
```

```python
s3uri_preprocessed_train = 's3://{}/{}'.format(bucket, preprocessed_test_path)
df_test = pd.read_csv(s3uri_preprocessed_train)
```

```python
df_test.head()
```

```python
df_test = df_test.drop(columns=['Survived'])
df_test.head()
```

```python
pred = predictor.predict(df_test)

print("Predictions:  \n", pred)
```

```python
df_test['Survived'] = pred
```

```python
df_test.head()
```

```python
df_submission = df_test[['PassengerId', 'Survived']]
```

```python
df_submission.head()
```

```python
df_submission['Survived'] = df_submission['Survived'].astype(int)
```

```python
df_submission.head()
```

```python
import boto3
```

```python
filename='predictions.csv'
```

```python
df_submission.to_csv(filename)
```

```python
submission_upload_path = 'submission/predictions.csv'
```

```python
s3 = boto3.resource('s3')
s3.meta.client.upload_file(filename, bucket, submission_upload_path)
```
